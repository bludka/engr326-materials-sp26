$$ y(t_{i+1}) = y(t_i) + h y'(t_i)$$

$$y(t_{i+1}) = y(t_i) + h y'(t_i) +\frac{h^2}{2} y''(t_i) + \frac{h^3}{3!} y^{(3)}(t_i) + \frac{h^4}{4!} y_i^{(4)}(t_i) +....$$

# 2. Local truncation error

If we only include $n$ terms in the Taylor Series, we can group the remaining terms in a remainder $R_n$

$$y(t_{i+1}) = y(t_i) + h y'(t_i) +\frac{h^2}{2} y''(t_i) + \frac{h^3}{3!} y^{(3)}(t_i) + \frac{h^4}{4!} y_i^{(4)}(t_i) + R_n$$

where 

$$R_n = h^{n+1} \frac{y^{(n+1)}(\xi)}{(n+1)!}$$

and $\xi$ is somewhere on the interval between $t_i$ and $t_i + h$.

Note that $y^{(n+1)}(\xi)$ means that $y^{(n+1)}$ is a function of $\xi$.
